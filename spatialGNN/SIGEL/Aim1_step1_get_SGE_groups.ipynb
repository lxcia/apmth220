## Get cluster centroids

```python
SGRs.shape
```

```python
SGRs_list_of_lists = [np.array(SGRs[i, :].cpu()) for i in range(SGRs.shape[0])]
```

```python
SGRs_df = pd.DataFrame({"SGR": SGRs_list_of_lists, "label": labels})
```

```python
group_centroids = SGRs_df.groupby("label").mean()
group_centroids.columns = ["centroid"]
```
